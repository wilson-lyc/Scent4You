```python
import pandas as pd
```

```python
note_class = pd.read_csv("data/note_miss_class_filled.csv")
```

```python
df = pd.read_csv("data/note_count_class.csv")
note_class_unique = note_class.drop_duplicates(subset='note', keep='first')
mask = df['class'].isna() | (df['class'] == '')
df.loc[mask, 'class'] = df.loc[mask, 'note'].map(note_class_unique.set_index('note')['class'])
```

```python
df.to_csv("data/note_class_map_raw.csv", index=False)
```
